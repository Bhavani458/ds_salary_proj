### Import libraries

```python
import pandas as pd
```

#### Loading csv file

```python
df= pd.read_csv('ds_salary_data.csv',index_col=0)
```

```python
df.head(100)
```

```python
# Data Cleaning to be done on:

# salary_estimate parsing
# state in location column
# company name text only
# company age (founded column)
# keywords in job description (python, R etc)
```

### Salary Estimate Parsing

```python
df.describe()
```

```python
df.info()
```

```python
df.shape
```

```python
# we need to drop these rows as they are not useful 
df[df['Salary Estimate'] == 'Easy Apply']
```

```python
df = df[df['Salary Estimate'] != 'Easy Apply']
```

```python
# now we have 427 rows
df.shape
```

```python
df.head()
```

```python
mask = df['Salary Estimate'].str.contains('years')
mask.value_counts()
#now drop all the values where mask value is 'True'
df.drop(df[mask].index,inplace=True)
```

```python
#split salary column and only take numeric range
salary = df['Salary Estimate'].apply(lambda x :x.split('(')[0])
salary
```

```python
minus_TL = salary.apply(lambda x :x.replace('T','').replace('L','00').replace('$',''))
minus_hr = minus_TL.apply(lambda x:x.lower().replace('per hour',''))
```

```python
#there are 25 hourly salary estimates
#create a new column with hourly values
df['Hourly']=df['Salary Estimate'].apply(lambda x: 1 if 'per hour' in x.lower() else 0)
df['Hourly'].value_counts()
```

```python
#create a column 'min_salary' and 'max_salary' convert the values to int

def extract_min_salary(value):
    try:
        return int(float(value.split('-')[0].strip()))
    except (ValueError, IndexError):
        return None

def extract_max_salary(value):
    try:
        return int(float(value.split('-')[1].strip()))
    except (ValueError, IndexError):
        return None

# Create 'min_salary' column
df['min_salary'] = minus_hr.apply(extract_min_salary)

# Create 'max_salary' column
df['max_salary'] = minus_hr.apply(extract_max_salary)
df['min_salary'].dtype
```

```python
#create a column for 'avg_salary'
df['avg_salary']=(df.min_salary+df.max_salary)/2
```

```python
df.head(10)
```

### Company Name Column

```python
#Only company name is kept
df['Company_text'] = df.apply(lambda x:x['Column Name'] if x['Rating']<0 else x['Company Name'].split('\n')[0],axis=1)
```

```python
df.head(10)
```

### State in location column

```python
#only state 
def extract_state(location):
    parts = location.split(',')
    if len(parts) > 1:
        return parts[1].strip()
    else:
        return None
df['job_state']=df['Location'].apply(extract_state)
df.job_state.value_counts()
```

### Company age (founded column)

```python
df['Company_age']=df.Founded.apply(lambda x:x if x<1 else 2023-x)
df.head()
```

### keywords in job description (python, R etc)

```python
df.reset_index(drop=True,inplace=True)
```

```python
#python
df['python_yn']=df['Job Description'].apply(lambda x:1 if 'python' in x.lower() else 0)
df['python_yn'].value_counts()
```

```python
#r-studio
df['r_studio']=df['Job Description'].apply(lambda x:1 if 'r studio' or 'r-studio' in x.lower() else 0)
df['r_studio'].value_counts()
```

```python
#spark
df['spark']=df['Job Description'].apply(lambda x:1 if 'spark' in x.lower() else 0)
df['spark'].value_counts()
```

```python
#aws
df['aws']=df['Job Description'].apply(lambda x:1 if 'aws' in x.lower() else 0)
df['aws'].value_counts()
```

```python
#excel
df['excel']=df['Job Description'].apply(lambda x:1 if 'excel' in x.lower() else 0)
df['excel'].value_counts()
```

```python
#sql
df['sql']=df['Job Description'].apply(lambda x:1 if 'sql' in x.lower() else 0)
df['sql'].value_counts()
```

```python
df_out = df
df_out.head()
```

```python
df_out.to_csv('Salary_data_cleaned.csv',index=False)
```
